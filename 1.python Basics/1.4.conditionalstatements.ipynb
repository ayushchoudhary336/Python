```python
## nested conditional statements

num=int(input("Enter the number : "))

if num>=0:
  print("the number is positive")
  if num%2==0:
    print("the number is even")
  else:
    print("the number is odd") 

else:
  print("the number is zero or negative")
```

```python
## practical Example

## you need to determine if a year is  a leap year using nested conditional statement


year=int(input("Enter the year : "))

if year%4==0:
  if year%100==0:
    if year%400==0:
      print(year, " is a leap year")
    else:
      print(year,"is not a leap year")  

  else:
    print(year,"is a leap yaer")    

else:
  print(year,"is not a leap year")    




```

While loop


```python
## the while loop continues to execute as long as the condition is true

count = 0

while count<5:
  print(count)
  count+=1
```

```python
## loop control statements

## break
## the break statemnets exits the loop prematurely

for i in range(10):
  if(i==5):
    break
  print(i)
```

```python
## continue

## the continue statements skips the current iteration and continues with the next.

for i in range(20):
  if (i%2!=0):
    continue
  print(i)
```

```python
## pass
## the pass statement is a null operation it does nothing.


for i in range(5):
  if i==3:
    print(" the number is 3")
    pass
  print(i)
```

```python
## nested loops
## a lopps inside loop

for i in range(3):
  for j in range(2):
    print(f"i: {i} and j:{j}")
```

```python
## examples  -Calculate the sum of first N natural numbers using a while and for loop

n=int(input("Enter the number : "))
sum=0
count=1

while count<=n:
  sum=sum+count
  count=count+1

print("sum of first n natural numbers : ",sum)  
```

```python
## prime number between 1 and 100
for num in range(1, 101):
    if num > 1:
        for i in range(2, num):
            if num % i == 0:
                break
        else:
            print(num)
```
