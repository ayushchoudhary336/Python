Tuples are immutable , meaning their elements cannot be changed once assigned


```python
## packing and unpacking tuple

packed_tuple=1,"hello",3.14
print(packed_tuple)
```

```python
## unpackaging
a,b,c=packed_tuple
print(a)
print(b)
print(c)
```

```python
## unpackaging with *

numbers=(1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```
