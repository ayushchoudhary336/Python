Dictionaries


Dictionaries are unordered collections of items .They store data in key-value pairs.Keys must be unique and immutable(e.g strings, numbers or tuples), while values can be of any type.


```python
## create Disctionaries
empty_dict={}
print(type(empty_dict))
```

```python
student={"name":"Ayush","age":21,"grade":"A"}
print(student)
print(type(student))
```

```python
## accessinf dictionaries elements by using keys
print(student["grade"])

## accessing using get() method
print(student.get('grade'))
print(student.get("last_name","Not Available"))
```

```python
student["age"]=35  ## update value
print(student)
student["country"]="India"  ## added a new key and value
print(student)
```

```python
del student["grade"]
print(student)
```

```python
## dictionaries methods

keys=student.keys() ## get all keys
print(keys)
values=student.values() ## get all values
print(values)

items=student.items() ## geta all key value pairs
print(items)
```

```python
## shallow copy 
student
```

```python
student_copy=student
print(student)
print(student_copy)
```

```python
student["name"]="Rohit"
print(student)
print(student_copy)
```

```python
## shallow copy
student_copy1=student.copy()
print(student_copy1)
print(student)
```

```python
student["age"]=40
print(student_copy1)
print(student)
```

```python
## iterating over dictionaries
## you can use loops to iterate over dictionaries , keys, values, or items

## iterating over keys

for keys in student.keys():
  print(keys)

## iterate  over values

for value in student.values():
  print(value)

## iterate over key, value pair

for key,value in student.items():
  print(f"{key}: {value}")


```

```python
## nested dictionaries

students={
  "student1":{"name":"Ayush","age":32},
  "student2":{"name":"Rahul","age":56}
}

students
```

```python
## Dictionary comprehension

squares={ x:x**2 for x in range(5)}
print(squares)

squareven={ x:x**2 for x in range(20) if x%2==0}
print(squareven)

```

```python
## practical Examples

## use a dictionary to count the frequency of elements in list

numbers=[1,2,2,3,3,3,4,4,4,4,4]
frequency={}

for number in numbers:
  if number in frequency:
    frequency[number]+=1
  else:
    frequency[number]=1

print(frequency)      
```

```python
## Merge 2 dictionaries into one

dict1={
  "a":1, "b":2
}
dict2={
  "c":3, "d":4
}

merged_dict={**dict1,**dict2}
print(merged_dict)
```
