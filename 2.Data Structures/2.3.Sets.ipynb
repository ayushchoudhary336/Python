Sets


```python
## create a set
my_set={1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set=set()
print(type(my_empty_set))
```

```python
## basic sets operation

## add and remove

my_set.add(7)
print(my_set)
```

```python
#remove
my_set.remove(3)
print(my_set)
```

```python
remove_element=my_set.pop()
print(remove_element)
print(my_set)
```

```python
# clear all the elements
my_set.clear()
print(my_set)
```

```python
## set Membership test
## whethe an element in present in set or not

my_set={1,2,3,4,5,7,8,9}
print(3 in my_set)
print(19 in my_set)
```

```python
## mathematical operation

set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

## union
union_set=set1.union(set2)
print(union_set)
```

```python
## intersection

intersection_set=set1.intersection(set2)
print(intersection_set)

set1.intersection_update(set2)
print(set1)
```
